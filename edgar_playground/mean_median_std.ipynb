```python
import numpy as np
import pandas as pd
np.random.seed(55)
```

```python
a = pd.DataFrame(np.round(np.random.rand(20,1) * 100 - 50))
a.columns = ['data']
a['cat'] = np.round(np.random.rand(20,1) * 3)
```

```python
median = a.groupby(a['cat'])[['data']].median().rename(columns={'data': 'median'})
mean = a.groupby(a['cat'])[['data']].mean().rename(columns={'data': 'mean'})
std = a.groupby(a['cat'])[['data']].std().rename(columns={'data': 'std'})

stat = pd.concat([mean, median, std], axis=1)

stat
```
